### insights from Data

Alcohol: Alcohol content affects a wine's body, since alcohol is more viscous than water. A wine with higher alcohol content will have a fuller, richer body, while a lower alcohol wine will taste lighter and more delicate on the palate. Wine type of Quality-1 has alcohol level of 13-15 and wine type Quality-2 has alcohol level of 11-13.5 and wine type Quality-3 has alcohol level of 12.5 to 14

Malic_Acid: Malic Acid is one of the main acids found in the acidity of grapes. Its concentration decreases the more a grape ripens. Malic Acid provides a strong link to wines tasting 'flat' if there is not enough. If there is too much the wine will taste 'sour'. so less Malic_acid means good quality wine. In the graph we can see that wine type Quality-1 has Malic_Acid in range of 1.5 to 4 and wine type Quality-2 has Malic_Acid in range of 1 to 4 and wine type Quality-3 has Malic_Acid in range of 1.5 to 5.6.

Ash:Ash content is one of the important indicators in wine quality determination.On the average about 2.5 g/L of ash are found in wine. Ash being defined as the inorganic matter that remains after evaporation and incineration. Cations - most of the ash falls into this class and includes potassium, sodium, calcium, magnesium, iron, copper, lead, arsenic, etc.In the graph  for wine type Quality-1 has Ash in range of 2 to 2.8 and wine type Quality-2 has Ash in range of 1.7 to 2.9 and wine type Quality-3 has Ash in Range of 2.0 to 2.8.

Alcalinity_of_ash:The alkalinity of the ash is defined as the sum of cations, other than the ammonium ion, combined with the organic acids in the wine. The alkalinity of ash will be expressed in milliequivalents per litre or in grams per litre of potassium carbonate.
in the graph for wine type Quality-1 has Alcalinity_of_ash in range of 11 to 22.5 and wine type Quality-2 has Alcalinity_of_ash in range of 15 to 25 and wine type Quality-3 has Alcalinity_of_ash in range of 17.5 to 25

Magnesium: Magnesium has a beneficial effect on VHG fermentations as it protects the yeast cells against ethanol, osmotic and temperature stress.
in the graph for wine type Quality-1 has Magnesium in range of 90 to 130 and wine type Quality-2 has Magnesium in range of 80 to 110 and wine type Quality-3 has Magnesium in range of 80 to 120.

Total_phenols:Most of the phenols in wine come from the pulp, skin, seeds and stems of grapes. There are more phenols in red wines than in white wines, in part because white wines are made primarily just from the juice of grapes, while red wines are made with more extracted grape bits. phenolic compounds contribute to sensorial characteristics of wines and the total phenolic content is also a worldwide standardized indicator to estimate the state of the quality of wine, therefore, rapid and accurate determination of total phenolic content in wine is of great importance for controlling sensory attributes and market value or quality.
In the graph for wine type Quality-1 has Total_phenols in range of 2.2 to 3.5 and wine type Quality-2 has Total_phenols in range of 1.5 to 3.5 and wine type Quality-3 has Total_phenols in range of 1.0 to 2.7.

Flavanoids:Flavonoids are rich in antioxidant activity and can help your body ward off everyday toxins. Including more flavonoids in your diet is a great way to help your body stay healthy and potentially decrease your risk of some chronic health conditions. So High amount of Flavanoids indicates Good quality wine.
In the graph for wine type Quality-1 has Flavanoids in range of 2 to 4 and wine type Quality-2 has Flavanoids in range of 1 to 3.5 and wine type Quality-3 has Flavanoids in range of 0.3 to 1.5.

Nonflavanoid_phenols:The non-flavonoids_phenols include the stilbenoids such as resveratrol and phenolic acids such as benzoic, caffeic and cinnamic acids.
In the graph for wine type Quality-1 has Nonflavanoid_phenols in range of 0.2 to 0.5 and wine type Quality-2 has Nonflavanoid_phenols in range of 0.1 to 0.6 and wine type Quality-3 has Nonflavanoid_phenols in range of 0.2 to 0.6. so we can say that less amount of Nonflavanoid_phenols tends to good quality of wine.

Proanthocyanins:Proanthocyanidins play an important role in wine; with the capability to bind salivary proteins, these condensed tannins strongly influence the perceived astringency of the wine. These compounds are typically present in levels of 300mg/L1 in red wine, though enological processing can affect the final concentrations.
In the graph for wine type Quality-1 has Proanthocyanins in range of1.2 to 3.0 and wine type Quality-2 has Proanthocyanins in range of 0.5 to 3.5 and wine type Quality-3 has Proanthocyanins in range of 0.5 to 2.8.

Color_intensity:Wine Color Intensity – a simple measure of how dark the wine is using a summation of absorbance measurements in the violet, green and red areas of the visible spectrum.The color of the wine mainly depends on the color of the drupe of the grape variety. Since pigments are localized in the center of the grape drupe, not in the juice, the color of the wine depends on the method of vinification and the time the must is in contact with those skins, a process called maceration.Intensity relates to appearance and aroma. When evaluating appearance, intensity describes the concentration of color. The more concentrated and opaque a wine's color, the higher its intensity.
In the graph for wine type Quality-1 has Color_intensity in range of 3 to 9 and wine type Quality-2 has Color_intensity in range of 2 to 7 and wine type Quality-3 has Color_intensity in range of 4 to 11. so we can say that less amount of Color_intensity tends to good quality of wine.

Hue:hue of a wine can give you hints as to the quality or freshness of a wine. If the wine is bright, and has a blue, silver or green hue, and colors that lean towards the beginning of the list, generally the wine will still be young or at least not past its peak. If the wine is dull or hazy, brown in color or orange in hue. This can mean that the wine has reached or past its peak.
Wines with more red colored hue have a lower pH (high acidity). Wines with a violet colored hue range from around 3.4–3.6 pH (on average). Wines with a more blueish tint (almost like magenta) are over 3.6 pH and possibly closer to 4 (low acidity).
In the graph for wine type Quality-1 has Hue in range of0.8 to 1.3 and wine type Quality-2 has Hue in range of 0.7 to 1.4 and wine type Quality-3 has Hue in range of 0.5 to 1.0.

OD280/OD315_of_diluted_wines: This is a method for determining the protein concentration, which can determine the protein content of various wines. Factor and cluster analysis are applied to the known data set to determine the category to which the wine sample belongs.
In the graph for wine type Quality-1 has OD280/OD315_of_diluted_wines in range of 2.5 to 4.0 and wine type Quality-2 has OD280/OD315_of_diluted_wines in range of 1.6 to 3.7 and wine type Quality-3 has OD280/OD315_of_diluted_wines in range of 1.0 to 2.5.

Proline:Proline is typically the most abundant amino acid present in grape juice and wine. The amount present is influenced by viticultural and winemaking factors and can be of diagnostic importance.However, the method was robust for the analysis of fermentation samples and table wines.
In the graph for wine type Quality-1 has Proline in range of 700 to 1600 and wine type Quality-2 has Proline in range of 200 to 1000 and wine type Quality-3 has Proline in range of 400 to 900. so we can say that higher amount of Proline tends to good quality wine.

## Model Creation

```python
## Creating independent and dependent variable
X=data.iloc[:,1:]
y=data.wine
```

```python
## Balacing the data
from collections import Counter
from imblearn.over_sampling import SMOTE
sm=SMOTE()
print(Counter(y))
X_sm,y_sm=sm.fit_resample(X,y)
print(Counter(y_sm))
```

```python
## scaling the data as all features seems to be near to normal distribution
from sklearn.preprocessing import StandardScaler 
scalar = StandardScaler()## objet creation
X_scaled = scalar.fit_transform(X_sm)
```

```python
X_scaled
```

```python
### creating train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y_sm,test_size=0.20,random_state=42)
```

```python
## checking the train and test shape
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN1=KNeighborsClassifier() ## model object creation
KNN1.fit(X_train,y_train)  ## fitting the model
y_hat_knn=KNN1.predict(X_test) ## getting the predict from created model
```

```python
from sklearn.metrics import f1_score
f1_knn=f1_score(y_test,y_hat_knn,average='weighted')
f1_knn
```

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_hat=dt.predict(X_test)
```

```python
## Evalauting the model
from sklearn.metrics import accuracy_score,classification_report,f1_score
## Training score
y_train_predict=dt.predict(X_train)
cc_train=accuracy_score(y_train,y_train_predict)
cc_train
```

```python
print(classification_report(y_train,y_train_predict))
```

```python
## test acc
test_acc=accuracy_score(y_test,y_hat)
test_acc
```

## Hyperparameter Tuning for Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=200, stop=2000, num=10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 40, num=9)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

random_grid = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}

rf_clf1 = RandomForestClassifier(random_state=42)

rf_cv = RandomizedSearchCV(estimator=rf_clf1, scoring='f1',param_distributions=random_grid, n_iter=100, cv=3, 
                               verbose=2, random_state=42, n_jobs=-1)

rf_cv.fit(X_train, y_train)
rf_best_params = rf_cv.best_params_
print(f"Best paramters: {rf_best_params})")

```

```python
rf_clf2 = RandomForestClassifier(**rf_best_params)
rf_clf2.fit(X_train, y_train)
y_predict=rf_clf2.predict(X_test)
f1_score=f1_score(y_test,y_predict,average='weighted')
```
